# Ensemble Learning
Ensemble learning is a technique in which we train so many on our given dataset.
and the small model which train on the training dataset called WEAK MODEL.
Now whatever sample we get will predict there class using the combination of week models.
STRONG MODEL: these small model collectively represent a Strong Model.

Ensemble(group)learning techniques:
> Bagging

NOTE: Bagging all algos create week model and data created using bootstrap.

NOTE: In Bootstrap same number of sample come in the training of weak model and each weak model may contain duplicate sample manytime.
1. Randomforest regression
2. Randomforest classifier
3. Bagging Classifier
4. Bagging Regressor
   etc.
> Boosting

1. AdaBoost Classifier
2. AdaBoost Regressor
   etc.
> Stacking
1. Voting Classifier
2. Voting Regressor

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
iris=pd.read_csv("IRIS.csv")
X16=iris.iloc[:,:-1].values
y16=iris.iloc[:,-1].values
#scaling
sc=MinMaxScaler()
X_sc16=sc.fit_transform(X16)
X_train16,X_test16,y_train16,y_test16=train_test_split(X_sc16,y16)
Model16=RandomForestClassifier(n_estimators=50,max_depth=1)
Model16.fit(X_train16,y_train16)
Model16.score(X_train16,y_train16)
Model16.score(X_test16,y_test16)

sample16=[[1,2,4,3]]
print("actual sample :",sample16)
sample_sc16=sc.transform(sample16)
print("transformed sample :",sample_sc16)
print("prediction :",Model16.predict(sample_sc16))
print("probability :",Model16.predict_proba(sample16))

print("Training score :",Model16.score(X_train16,y_train16))
print("Testing score :",Model16.score(X_test16,y_test16))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
iris=pd.read_csv("IRIS.csv")
X16=iris.iloc[:,:-1].values
y16=iris.iloc[:,-1].values
#scaling
sc=MinMaxScaler()
X_sc16=sc.fit_transform(X16)

for i in range(10):
    X_train16,X_test16,y_train16,y_test16=train_test_split(X_sc16,y16,random_state=i)
    Model16=RandomForestClassifier(n_estimators=100,max_depth=2)
    Model16.fit(X_train16,y_train16)
    Model16.score(X_train16,y_train16)
    Model16.score(X_test16,y_test16)

    sample16=[[1,2,4,3]]
    print("actual sample :",sample16)
    sample_sc16=sc.transform(sample16)
    print("transformed sample :",sample_sc16)
    print("prediction :",Model16.predict(sample_sc16))
    print("probability :",Model16.predict_proba(sample16))

    print("Training score :",Model16.score(X_train16,y_train16))
    print("Testing score :",Model16.score(X_test16,y_test16))
    print(f"-------------------{i}-----------------------")

 
```
