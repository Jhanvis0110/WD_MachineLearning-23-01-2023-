# Disadvantage:

```python
from sklearn.datasets import load_iris
df=load_iris()
X=df['data']
y=df['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Model10=DecisionTreeClassifier()
Model10.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
from sklearn.tree import plot_tree

plot_tree(Model10)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("IRIS.csv")
X=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
from sklearn.tree import DecisionTreeClassifier
Model11=DecisionTreeClassifier()
Model11.fit(X,y)

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plot_tree(Model11)
plt.show()
```

```python
Sample=[[1.2,3,5,2]]
```

```python
Model11.predict(Sample)
```

```python
Model11.predict_proba(Sample)
```

```python
Model12=DecisionTreeClassifier(max_depth=3)
Model12.fit(X,y)
plot_tree(Model12)
plt.show()

```

```python
Model12.predict(Sample)
```

```python
Model12.predict_proba(Sample)
```
